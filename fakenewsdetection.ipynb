```python
import nltk
#nltk.download('all')
nltk.download('omw-1.4')

import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
stop_words = set(stopwords.words('english')) 
def LemmSentence(sentence):
    lemma_words = []
    wordnet_lemmatizer = WordNetLemmatizer()
    word_tokens = word_tokenize(sentence) 
    for word in word_tokens: 
        if word not in stop_words: 
            new_word = re.sub('[^a-zA-Z]', '',word)
            new_word = new_word.lower()
            new_word = wordnet_lemmatizer.lemmatize(new_word)
            lemma_words.append(new_word)
    return " ".join(lemma_words)

X = [LemmSentence(i) for i in X]
```

```python
X[1]
```

```python
X = pd.DataFrame(X)
y = pd.DataFrame(y)
X.shape, y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=7)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()

```

```python
# transforming
tfidf_train = vectorizer.fit_transform(x_train.iloc[:,0])
tfidf_test = vectorizer.transform(x_test.iloc[:,0])
```

```python
tfidf_train.shape
```

```python
from sklearn.linear_model import LogisticRegression

LR = LogisticRegression()
LR.fit(tfidf_train,y_train)
```

```python
pred_lr=LR.predict(tfidf_test)
LR.score(tfidf_test, y_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier

DT = DecisionTreeClassifier()
DT.fit(tfidf_train, y_train)

pred_dt = DT.predict(tfidf_test)
```

```python
print(classification_report(y_test, pred_dt))
```
